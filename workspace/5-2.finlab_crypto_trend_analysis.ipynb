```python
ohlcv = finlab_crypto.crawler.get_all_binance('BTCUSDT', '4h')
```

```python
from finlab_crypto import Strategy

@Strategy(filter_name='sma', n1=20, n2=60)
def trend_strategy(ohlcv):

  filter_name = trend_strategy.filter_name
  v1 = trend_strategy.n1
  v2 = trend_strategy.n2

  close = ohlcv.close
  sma20 = indicators.trends[filter_name](close, v1)
  sma60 = indicators.trends[filter_name](close, v2)

  entries = (sma20 > sma60) & (sma20.shift() < sma60.shift())
  exits = (sma20 < sma60) & (sma20.shift() > sma60.shift())

  return entries, exits

import numpy as np
variables = {
    #list(indicators.trends.keys())=['sma', 'wma', 'lowpass', 'hullma', 'zlma', 'alma', 'detrend', 'linear_reg'],
    'filter_name': list(indicators.trends.keys()),
    'n1': np.arange(20, 300, 20),
    'n2': np.arange(20, 300, 20)
}

# Diff:final_value() is replace to final_equity() in vectorbt==0.14.4,so the code is different from vedio.
portfolio = trend_strategy.backtest(ohlcv, variables=variables, freq='4h', plot=False)

```

# Bootstrap Test

## Detrend daily returns

[unstack api reference](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.unstack.html)

```python
# Diff:daily_returns is func in vectorbt==0.14.4, not method, so the code is different from vedio.
detrend = portfolio.daily_returns() - portfolio.daily_returns().mean()
samples = detrend.unstack().values
```

## Resample

[random.choice api reference](https://docs.python.org/3/library/random.html)

```python
import random
# Diff:daily_returns is func in vectorbt==0.14.4, not method, so the code is different from vedio.
ndays = len(portfolio.daily_returns())

random_returns = []
for i in range(1000):
  random_returns.append(sum(random.choices(samples, k=ndays)))

pd.Series(random_returns).hist()
```
